```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_performance.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df[(df['Attendance(%)'] >= 0) & (df['Attendance(%)'] <= 100)]
```

```python
df = df[(df['Study_Hours'] >= 0) & (df['Study_Hours'] <= 12)]
```

```python
print(df.shape)
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Attendance(%)', y='Test_Score', data=df)
plt.title("Attendance vs Test Score")
plt.xlabel("Attendance Percentage")
plt.ylabel("Test Score")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Study_Hours', y='Test_Score', data=df)
plt.title("Study Hours vs Test Score")
plt.xlabel("Study Hours")
plt.ylabel("Test Score")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='Grade', data=df)
plt.title("Grade Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

INSIGHTS

1. Study hours show a strong positive correlation with test scores.

2. Students who studied more hours generally achieved better academic performance.

3. Attendance percentage showed very weak correlation with test scores in this dataset.

4. Grade distribution indicates that many students scored lower grades.

5. Study habits appear to influence academic performance more than attendance in this analysis.
